```python
from just_d4rl import D4RLDataset, d4rl_offline_dataset, D4RLScoreNormalizer
import gymnasium as gym
import matplotlib.pyplot as plt
import torch
import numpy as np
import torch.nn as nn
from stable_baselines3.common import policies
from stable_baselines3.common.torch_layers import FlattenExtractor, MlpExtractor
import collections
from torch.utils.data import DataLoader, TensorDataset
from prettytable import PrettyTable
from scipy.ndimage import convolve1d
from scipy.ndimage import gaussian_filter1d
from scipy.signal.windows import triang

# Evaluation and Saving Functions
def evaluate_on_env(model, device, env, desired_return_target,
                    num_eval_ep=10, max_test_ep_len=1000,
                    state_mean=None, state_std=None, render=False,
                    state_normalization=True):
    eval_batch_size = 1
    results = {}
    total_reward = 0
    total_timesteps = 0

    state_dim = env.observation_space.shape[0]
    act_dim = env.action_space.shape[0]

    if state_mean is None:
        state_mean = torch.zeros((state_dim,)).to(device)
    else:
        state_mean = torch.from_numpy(state_mean).to(device)

    if state_std is None:
        state_std = torch.ones((state_dim,)).to(device)
    else:
        state_std = torch.from_numpy(state_std).to(device)

    model.eval()
    model.to(device)

    with torch.no_grad():
        return_per_ep_log = []
        all_rewards = []

        for ep in range(num_eval_ep):

            reset_result = env.reset()
            if isinstance(reset_result, tuple):  # Gymnasium style
                running_state, _ = reset_result
            else:  # Old Gym style
                running_state = reset_result

            running_reward = 0
            reward_per_ep = 0
            running_reward_log = []
            desired_return = desired_return_target
            for t in range(max_test_ep_len):
                total_timesteps += 1
                desired_return -= running_reward*0.001

                state = torch.from_numpy(running_state).to(device, dtype=torch.float32)
                if state_normalization:
                    state = (state - state_mean) / state_std

                if len(state.shape) == 1:
                    state = state.unsqueeze(0)

                state = torch.cat([state, torch.from_numpy(np.array(desired_return)).to(device, dtype=torch.float32).reshape((1,1))], dim=1)

                actions = model._predict(state, deterministic=True)

                step_result = env.step(actions.cpu().numpy().flatten())
                if len(step_result) == 5:  # Gymnasium style
                    running_state, running_reward, terminated, truncated, _ = step_result
                    done = terminated or truncated
                else:  # Old Gym style
                    running_state, running_reward, done, _ = step_result

                running_reward_log.append(running_reward)

                total_reward += running_reward
                reward_per_ep += running_reward

                if render:
                    env.render()
                if done:
                    print('sum of reward per episode: ', reward_per_ep)
                    return_per_ep_log.append(reward_per_ep)
                    break

    t = np.arange(len(running_reward_log))
    plt.plot(t, np.array(running_reward_log), 'b')
    plt.show()

    results['eval/avg_reward'] = total_reward / num_eval_ep
    results['eval/avg_ep_len'] = total_timesteps / num_eval_ep
    results['eval/std_returns'] = np.std(return_per_ep_log)
    results['eval/all_rewards'] = all_rewards

    return results

def save_model(model, dataset_name):
    model_filename = f"trained_mle_model_{dataset_name}.pth"
    torch.save(model.state_dict(), model_filename)
    print(f"Model saved to {model_filename}")

# Data Processing Functions
def flatten_traj(paths):
    states_list = []
    actions_list = []

    for p in paths:
        states_list.append(p['observations'])
        actions_list.append(p['actions'])

    states = np.concatenate(states_list)
    actions = np.concatenate(actions_list)

    return(states, actions)

def weighted_mse_loss(inputs, targets, weights=None):
    loss = (inputs - targets) ** 2
    if weights is not None:
        loss *= weights.expand_as(loss)
    loss = torch.mean(loss)
    return loss

def get_lds_kernel_window(kernel, ks, sigma):
    assert kernel in ['gaussian', 'triang', 'laplace']
    half_ks = (ks - 1) // 2
    if kernel == 'gaussian':
        base_kernel = [0.] * half_ks + [1.] + [0.] * half_ks
        kernel_window = gaussian_filter1d(base_kernel, sigma=sigma) / max(gaussian_filter1d(base_kernel, sigma=sigma))
    elif kernel == 'triang':
        kernel_window = triang(ks)
    else:
        laplace = lambda x: np.exp(-abs(x) / sigma) / (2. * sigma)
        kernel_window = list(map(laplace, np.arange(-half_ks, half_ks + 1))) / max(map(laplace, np.arange(-half_ks, half_ks + 1)))
    return kernel_window

def apply_lds_weights_multivariate(targets, max_target, kernel, ks, sigma):
    weights_per_feature = []
    for i in range(targets.shape[1]):
        value_dict = {x: 0 for x in range(max_target)}
        for label in targets[:, i]:
            # Clip the label to ensure it falls within the valid range
            clipped_label = max(0, min(max_target - 1, int(label)))
            value_dict[clipped_label] += 1
        kernel_window = get_lds_kernel_window(kernel, ks, sigma)
        smoothed_values = convolve1d(
            np.asarray([v for _, v in value_dict.items()]), weights=kernel_window, mode='constant')
        weights = [1 / smoothed_values[max(0, min(max_target - 1, int(label)))] for label in targets[:, i]]
        scaling = len(weights) / np.sum(weights)
        weights = [scaling * x for x in weights]
        weights_per_feature.append(np.array(weights, dtype=np.float32))
    return np.mean(weights_per_feature, axis=0)

def mle_loss(policy, obs, actions, weights=None):
    _, log_prob, _ = policy.evaluate_actions(obs, actions)
    loss = -log_prob
    if weights is not None:
        loss *= weights
    return loss.mean()

# Policy Class
class UnconditionalPolicy(policies.ActorCriticPolicy):
    def __init__(
        self,
        observation_space,
        action_space,
        hidden_size=1024,
        depth=2,
        learning_rate=1e-3,
        dropout_p=0.1
    ):
        net_arch = dict(
            pi=[hidden_size] * depth,  # policy network architecture
            vf=[64]  # Small value network (we don't use it but need it)
        )

        super().__init__(
            observation_space=observation_space,
            action_space=action_space,
            lr_schedule=lambda _: learning_rate,
            net_arch=net_arch,
            activation_fn=nn.ReLU,
            features_extractor_class=FlattenExtractor,
            optimizer_class=torch.optim.Adam
        )

    def _build(self, lr_schedule):
        self.features_extractor = self.features_extractor_class(self.observation_space)
        self.features_dim = self.features_extractor.features_dim

        self.mlp_extractor = MlpExtractor(
            self.features_dim,
            net_arch=self.net_arch,
            activation_fn=self.activation_fn,
        )

        self.value_net = nn.Linear(self.mlp_extractor.latent_dim_vf, 1)

        latent_dim_pi = self.mlp_extractor.latent_dim_pi
        self.action_net, self.log_std = self.action_dist.proba_distribution_net(
            latent_dim=latent_dim_pi,
            log_std_init=self.log_std_init
        )

# Main Script

dataset_list = [
    ("halfcheetah-medium-replay-v2", 12000*0.001),
    ("hopper-medium-replay-v2", 3600*0.001),
    ("walker2d-medium-replay-v2", 5000*0.001)
]

results_table = PrettyTable()
results_table.field_names = ["Dataset", "Mean Last 10 Average Scores", "Mean Last 10 Std Scores"]

lr = 5e-4
weight_decay = 1e-4
epochs = 100
device = 'cuda'

for dataset_name, desired_return in dataset_list:
    print(f"Training on dataset: {dataset_name}")

    eval_env = gym.make(dataset_name.split('-')[0].replace('halfcheetah', 'HalfCheetah').replace('hopper', 'Hopper').replace('walker2d', 'Walker2d') + '-v4')

    dataset = d4rl_offline_dataset(dataset_name)
    get_normalized_score = D4RLScoreNormalizer(dataset_name)

    states = dataset['observations']
    actions = dataset['actions']

    act_dim = actions.shape[1]
    obs_dim = states.shape[1]

    paths = []
    N = dataset['rewards'].shape[0]
    data_ = collections.defaultdict(list)
    use_timeouts = 'timeouts' in dataset
    episode_step = 0

    for i in range(N):
        done_bool = bool(dataset['terminals'][i])
        final_timestep = dataset['timeouts'][i] if use_timeouts else (episode_step == 1000 - 1)

        for k in ['observations', 'next_observations', 'actions', 'rewards', 'terminals']:
            data_[k].append(dataset[k][i])

        if done_bool or final_timestep:
            episode_data = {k: np.array(data_[k]) for k in data_}
            paths.append(episode_data)
            data_ = collections.defaultdict(list)
            episode_step = 0
        episode_step += 1

    gamma = 1.0
    returns_input = []

    for path in paths:
        return_sum = 0
        returns_per_traj = np.zeros_like(path['rewards'])
        for t in reversed(range(len(path['rewards']))):
            return_sum = gamma * return_sum + path['rewards'][t] * 0.001
            returns_per_traj[t] = return_sum
        returns_input.append(returns_per_traj)
    returns_input = np.concatenate(returns_input)

    states_input, actions_input = flatten_traj(paths)
    weights = apply_lds_weights_multivariate(actions_input, max_target=51, kernel='gaussian', ks=9, sigma=1.0)

    states_mean = states_input.mean(axis=0)
    states_std = states_input.std(axis=0) + 1e-6
    normalized_states = (states_input - states_mean) / states_std

    states_tensor = torch.tensor(normalized_states, dtype=torch.float32).to(device)
    actions_tensor = torch.tensor(actions_input, dtype=torch.float32).to(device)
    weights_tensor = torch.tensor(weights, dtype=torch.float32).to(device)

    augmented_obs_space = gym.spaces.Box(
        low=np.concatenate([eval_env.observation_space.low, [-np.inf]]),
        high=np.concatenate([eval_env.observation_space.high, [np.inf]]),
        shape=(eval_env.observation_space.shape[0] + 1,),
        dtype=np.float32
    )

    actor = UnconditionalPolicy(
        observation_space=augmented_obs_space,
        action_space=eval_env.action_space,
        hidden_size=1024,
        depth=2
    )

    optimizer = torch.optim.Adam(actor.parameters(), lr=lr, weight_decay=weight_decay)

    states_tensor_with_returns = torch.cat((states_tensor, torch.tensor(returns_input, dtype=torch.float32).to(device).unsqueeze(1)), dim=1)

    train_loader = DataLoader(TensorDataset(states_tensor_with_returns, actions_tensor, weights_tensor), batch_size=512, shuffle=True)

    actor.to(device)

    train_losses = []
    eval_scores = []
    eval_stds = []
    max_eval_score = float('-inf')

    for epoch in range(epochs):
        actor.train()
        epoch_loss = 0.0
        for x_batch, y_batch, w_batch in train_loader:
            x_batch, y_batch, w_batch = x_batch.to(device), y_batch.to(device), w_batch.to(device)
            optimizer.zero_grad()

            loss = mle_loss(actor, x_batch, y_batch, weights=w_batch)
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        print('Epoch loss: ', epoch_loss)
        train_losses.append(epoch_loss / len(train_loader))

        if epoch % 5 == 0:
            results = evaluate_on_env(actor, device, eval_env, desired_return, num_eval_ep=5, max_test_ep_len=1000,
                                      state_mean=states_mean, state_std=states_std, render=False, state_normalization=True)
            eval_d4rl_score = get_normalized_score(results['eval/avg_reward']) * 100
            eval_d4rl_std = get_normalized_score(results['eval/std_returns']) * 100
            eval_scores.append(eval_d4rl_score)
            eval_stds.append(eval_d4rl_std)

            print(f'Epoch {epoch}, Average score: {eval_d4rl_score}')
            print(f'Std score: {eval_d4rl_std}')

            max_eval_score = max(max_eval_score, eval_d4rl_score)

    save_model(actor, dataset_name)

    avg_train_loss = sum(train_losses) / len(train_losses)
    mean_last_10_scores = np.mean(eval_scores[-10:])
    mean_last_10_stds = np.mean(eval_stds[-10:])

    print(f"\nMean of the last 10 Average Scores for {dataset_name}: {mean_last_10_scores}")
    print(f"Mean of the last 10 Std Scores for {dataset_name}: {mean_last_10_stds}")

    results_table.add_row([dataset_name, mean_last_10_scores, mean_last_10_stds])

print("\nFinal Results Table:")
print(results_table)
```
